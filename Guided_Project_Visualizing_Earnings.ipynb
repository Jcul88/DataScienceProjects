```python
recent_grads.plot(x="Full_time", y="Median", kind='scatter')
```

From the figure above, Median salary of full-time year-round workers decreases as the number of female graduate increases. Suggesting that a large number of female graduates earn less compared to a just a few.

```python
from pandas.plotting import scatter_matrix#imports scatter matrix 
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

Because scatter matrix plots are frequently used in the exploratory data analysis, pandas contains a function named scatter_matrix() that generates the plots for us. This function is part of the pandas.plotting module and needs to be imported separately.

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen')
recent_grads[162:].plot.bar(x='Major', y='ShareWomen')
```

Result from the first ten rows indicates high proportion women graduating with Astronomy & Astrophysics compared to percentage of women in other Majors.

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate')
recent_grads[162:].plot.bar(x='Major', y='Unemployment_rate')
```

Bar plots compare the unemployment rate (Unemployment_rate) between Majors from the first and last ten rows. Clinical Psychology Major has the highest unemployment rate for the last ten data points, while Early Childhood Education has the lowest unemployment rate among the last ten data points.

For the first ten rows, Nuclear Engineering Major has the highest unemployment rate, while Petroleum Engineering, Metallurgical Engineering and Astronomy & Astrophysics have the lowest unemployment rates.
